```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('./results/PBMC/test_unpaired/test_unpaired_v6/models_comparison_results - check_PBMC_v6.csv')
df.head()
```

```python
df_asw = pd.read_csv('./results/PBMC/test_unpaired/test_unpaired_v6/models_comparison_results - ASW_unpaired.csv')
df.head()
```

```python

mean_recall = df.groupby(['Models', 'k'])['Recall_at_k'].mean().reset_index()

plt.figure(figsize=(10, 8))
sns.lineplot(
    data=mean_recall,
    x="k",
    y="Recall_at_k",
    hue="Models",
    style="Models",
    markers=True,
    dashes=False
)

plt.xlabel("k", fontsize=10)
plt.ylabel("Recall@k", fontsize=10)
plt.xticks([10, 20, 30, 40, 50])
plt.legend(title="Models", fontsize=10, title_fontsize=12)
plt.tight_layout()
plt.show()
```

```python
mod_colors = {
    "Modality 1": "#7fc97f",  # Green
    "Modality 2": "#beaed4",  # Purple
    "Joint": "#fdc086"        # Orange
}

# Create the grouped box plot
plt.figure(figsize=(12, 6))
sns.boxplot(
    x="prop", 
    y="asw", 
    hue="mod", 
    data=df, 
    palette=mod_colors, 
    hue_order=["Modality 1", "Modality 2", "Joint"]
)

# Add labels and title
plt.xlabel("Proportion of Data", fontsize=12)
plt.ylabel("ASW (Adjusted Silhouette Width)", fontsize=12)
plt.title("Grouped Box Plot of ASW by Proportion and Modality", fontsize=14)

# Add legend
plt.legend(title="Modality", fontsize=10, title_fontsize=12)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
